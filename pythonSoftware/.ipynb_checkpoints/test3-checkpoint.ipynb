```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense,LSTM,Dropout, Activation

import datetime
```

```python
dataset = pd.read_csv('../dataset/005930.KS.csv', sep=',')

data = dataset.dropna()

dataset.head()
dataset.info()
dataset.describe()
```

```python
df=data[['Close']]
df.shape
df=np.array(df)
```

```python
sequen_len=50
seq_len =sequen_len+1


windows=[]
x=[]
y=[]
for i in range(len(df)-seq_len):
    windows.append(df[i:i+seq_len])

#windows = np.array(windows)
```

```python
normalized_data = []
for win in windows:
    normalized_window = [(float(p) - float(min(win))) / (float(max(win)) - float(min(win))) for p in win]
    normalized_data.append(normalized_window)
```

```python
# normalized_data = []
# for win in windows:
#     normalized_window = [(p - min(win)) / (max(win) - min(win)) for p in win]
#     normalized_data.append(normalized_window)
```

```python
# from sklearn.preprocessing import MinMaxScaler

# normalized_data = []

# scaler = MinMaxScaler()

# for win in windows:
#     normalized_window = scaler.fit_transform(win)
#     normalized_data.append(normalized_window)

# normalized_data[0]
```

```python
windows = np.array(normalized_data)

x = windows[:,:-1]
x = np.reshape(x,(x.shape[0], x.shape[1],1))
y = windows[:,-1]


#x = np.reshape(x,(x.shape[0], x.shape[1],1))'''

```

```python
seed = 0
np.random.seed(seed)
tf.random.set_seed(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=seed)
```

```python
model = Sequential()
model.add(LSTM(60, return_sequences=True, input_shape=(x.shape[1],x.shape[2])))
model.add(LSTM(50, return_sequences=True))
model.add(LSTM(32, return_sequences=True))
model.add(LSTM(20, return_sequences=False))
model.add(Dense(1, activation='linear'))

model.compile(loss='mse', optimizer='rmsprop')
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
es = EarlyStopping(monitor='val_loss', patience=5)
mc = ModelCheckpoint('best_model.h5', monitor='val_loss', mode='auto', verbose=1, save_best_only=True)
```

```python
history=model.fit(X_train, Y_train, validation_data=(X_test, Y_test), callbacks=[es, mc], batch_size=10, epochs=1000)
```

```python
print(history.history['acc'])

score = model.evaluate(X_test, Y_test, verbose=1)
print(score)
```

```python
epochs = range(1, len(history.history['val_loss']) + 1)
plt.plot(epochs, history.history['loss'])
plt.plot(epochs, history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
Y_prediction = model.predict(X_test)

fig = plt.figure(facecolor='white', figsize=(20, 10))
ax = fig.add_subplot(111)
ax.plot(Y_test, label='True')
ax.plot(Y_prediction, label='Prediction')
ax.legend()
plt.show()
```
